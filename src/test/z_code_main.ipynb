```python
import os
import sys

src_dir = os.path.dirname(os.getcwd())
project_dir = os.path.dirname(src_dir)
sys.path.append(project_dir)

import json
from datetime import date

from src.utils.helper import get_monday_of_week

PARTITION_DATE = get_monday_of_week().strftime("%Y-%m-%d")
TODAY = date.today().strftime("%Y-%m-%d")

def deep_get(d, path: list, default=None):
    """Truy cập an toàn vào dict lồng nhau."""
    for key in path:
        if isinstance(d, dict):
            d = d.get(key, default)
        else:
            return default
    return d

def count_json_types(folder_path: str):
    count_type1 = 0
    count_type2 = 0
    total_files = 0
    error_files = 0
    test = []
    for root, _, files in os.walk(folder_path):
        for f in files:
            if not f.endswith(".json"):
                continue

            file_path = os.path.join(root, f)
            total_files += 1

            try:
                with open(file_path, "r", encoding="utf-8") as infile:
                    data = json.load(infile)

                # --- Kiểm tra loại 1 ---
                item_struct = deep_get(data, ["data", "__DEFAULT_SCOPE__", "webapp.video-detail", "itemInfo", "itemStruct"])
                if isinstance(item_struct, dict) and "serverABVersions" in item_struct:
                    count_type1 += 1
                    # os.remove(file_path)
                    print(f"Loại 1: {file_path}")
                    continue  # tránh đếm lại

                # --- Kiểm tra loại 2 ---
                video_detail = deep_get(data, ["data", "__DEFAULT_SCOPE__", "webapp.video-detail"])
                if isinstance(video_detail, dict):
                    keys = set(video_detail.keys())
                    if keys == {"statusCode", "statusMsg"} and video_detail.get("statusCode", 0) != 0:
                        count_type2 += 1
                        print(f"Loại 2: {file_path}")
                        test.append(f.split(".json")[0])
                        continue

            except Exception as e:
                error_files += 1
                print(f"Lỗi khi đọc {file_path}: {e}")

    print(f"Tổng số file JSON: {total_files}")
    print(f"Số file lỗi không đọc được: {error_files}")
    print(f"Loại 1 (có serverABVersions): {count_type1}")
    print(f"Loại 2 (statusCode != 0): {count_type2}")
    print(f"Các file khác không thuộc 2 loại trên: {total_files - error_files - count_type1 - count_type2}")
    return test
# Ví dụ chạy:
test = count_json_types(fr"./data_tiktok_video/{PARTITION_DATE}/{TODAY}/raw_data/video")
print(test)
```

```python
import os
import json

def check_share_url_vs_aweme_id(folder_path: str):
    """Duyệt qua tất cả file JSON trong folder, kiểm tra share_url có chứa aweme_id không."""
    for filename in os.listdir(folder_path):
        if not filename.endswith(".json"):
            continue

        file_path = os.path.join(folder_path, filename)
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                data = json.load(f)

            share_url = data.get("share_url", "")
            aweme_id = str(data.get("aweme_id", ""))

            # Kiểm tra xem aweme_id có nằm trong share_url không
            if aweme_id not in share_url:
                print(f"❌ File: {filename}")
                print(f"    share_url: {share_url}")
                print(f"    aweme_id : {aweme_id}\n")

        except Exception as e:
            print(f"⚠️ Lỗi khi đọc file {filename}: {e}")

# Ví dụ chạy
folder_path = fr"./data_tiktok_video/{PARTITION_DATE}/{TODAY}/transform_data" 
check_share_url_vs_aweme_id(folder_path)
```
